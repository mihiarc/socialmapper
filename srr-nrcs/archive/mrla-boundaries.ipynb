```python
import geopandas as gpd
import contextily as ctx
import pandas as pd

import matplotlib.pyplot as plt

# Read the shapefile
folder_path = '/Users/mihiarc/Work/data/spatial-boundaries'
shapefile_path = folder_path + '/MLRA_52_2022/MLRA_52.shp'
gdf = gpd.read_file(shapefile_path)
# load state boundaries
state_boundaries = gpd.read_file(folder_path + '/tl_2023_us_state/tl_2023_us_state.shp')
# load county boundaries
county_boundaries = gpd.read_file(folder_path + '/tl_2023_us_county/tl_2023_us_county.shp')

# List of counties
counties = [
    'Wallace, KS', 'Logan, KS', 'Gove, KS', 'Lane, KS', 'Hamilton, KS', 'Kearny, KS',
    'Morton, KS', 'Graham, KS', 'Trego, KS', 'Ellis, KS', 'Ness, KS', 'Rush, KS',
    'Hodgeman, KS', 'Ford, KS', 'Lea, NM', 'Roosevelt, NM', 'Andrews, TX',
    'Meade, KS', 'Beaver, OK', 'Ellis, OK', 'Lipscomb, TX', 'Roberts, TX', 'Hemphill, TX',
    'Gray, TX'
]

fips = [
    '20199', '20109', '20063', '20101', '20075', '20093',
    '20129', '20065', '20195', '20051', '20135', '20165',
    '20083', '20057', '35025', '35041', '48003',
    '20119', '40007', '40045', '48295', '48395', '48211',
    '48179'
]
# select MRLAs of interest 72, 73, 77D, 77E
gdf = gdf[gdf['MLRARSYM'].isin(['72', '73', '77D', '77E'])]
# select the counties of interest
county_boundaries = county_boundaries[county_boundaries['GEOID'].isin(fips)]
# set the crs of the county boundaries to MRLA
county_boundaries = county_boundaries.to_crs(gdf.crs)
```
